# Conversion to gray scale 

```python
#Conversion is done based on the standard color to grayscale 
#conversion formula on Wikipedia i.e Y'=0.2627R'+ 0.6780G'+ 0.0593B'
def convert_to_greyScale(image):
    shape = image.shape
    
    #check if image is in rgb format
    if len(shape)==3:
            r, g, b = image[:,:,0], image[:,:,1], image[:,:,2]
            gray_img = 0.2989*r + 0.5870*g + 0.1140*b
            return gray_img
    
    else:
        return image
```


# Image Blurr 

```python
def smooth_img(img):
    row,col = img.shape
    img_blur = np.zeros((row,col))
    sigma=1
    size=2
    #initialize the parameters for slicing
    row_slice = 5
    
    #gaussian filter of 5x5
    x, y = np.mgrid[-size:size+1, -size:size+1]
    normal = 1 / (2.0 * np.pi * sigma**2)
    gf =  np.exp(-((x**2 + y**2) / (2.0*sigma**2))) * normal
    
    #convolve with the image but no need to
    #flip it because it is symmetric
    #1:Slice the image according to filter size
       #1:Slice the image according to filter size
    for i in range(row-4):
        col_slice= 5
        
        for j in range(col-4):
            img_slice = img[i:row_slice , j:col_slice]
            
            #2: convole and add the value into the corresonding derivative images
            img_blur[i+1, j+1] = sum(sum(img_slice*gf))
            
            #increment column slice size by 1
            col_slice = col_slice+1
            
        #increment the row slice by 1
        row_slice = row_slice+1
        
    return img_blur
```

# Mask generation

```python
def mask_generation(sigma):
    T = 0.3
    scale = 255
    Gfx = []
    Gfy = []
    
    #1: calculate filter size
    filter_size, sHalf = calculate_filter_size(sigma,T)
    
    #2: Generate filters
    x = np.arange(-sHalf,sHalf+1)
    y = np.arange(-sHalf,sHalf+1)
    X,Y = np.meshgrid(x,y)
    
    #3: generate gaussian kernal
    X_flatten = X.flatten()
    Y_flatten = Y.flatten()
    
    #calculate gradiant filters
    for x,y in zip(X_flatten, Y_flatten):
        fx,fy = calculate_gradiant(sigma,x,y)
        Gfx.append(fx*scale)
        Gfy.append(fy*scale)
    
    #4: Gx Gy are in list form, convert into matrix of filter_size x filter_size
    Gx = np.reshape(np.rint((np.array(Gfx))),(filter_size,filter_size))
    Gy = np.reshape(np.rint((np.array(Gfy))),(filter_size,filter_size))    
    #print(X,'\n\n',Y,"\n\n",Gx,"\n\n",Gy)
    return Gx,Gy
```

```python
#Calculate the gradiant of gaussian
def calculate_gradiant(sigma,x,y):
    #Given gaussian formula ----> G(x,y)=exp(-(x^2+y^2)/(2*sigma^2))
    #first derivative of above formula w.r.t x and  y
    fx = m.exp(-(x**2 + y**2)/(2*sigma**2)) * (x/(sigma**2))
    fy = m.exp(-(x**2 + y**2)/(2*sigma**2)) * (y/(sigma**2))
    return fx, fy
```

```python
#function for calculating filter size
def calculate_filter_size(sigma,T):
    sHalf = round(m.sqrt(-m.log(T)*2*sigma**2))
    N = 2* sHalf+1
    return N, sHalf
```

# Convolution

```python
#convolution of gray scale Circle imgae with Gx and Gy

def convolution(img, grad_x, grad_y):
    #print(grad_x,'\n\n', grad_y)
    fx = np.zeros((img.shape))
    fy = np.zeros((img.shape))
    
    #get the size of filter to determine the range for traverse
    filter_size = grad_x.shape
    
    row,col = img.shape
    row = row-filter_size[0]+1
    col = col-filter_size[1]+1
    
    grad_x_flip = np.zeros((grad_x.shape))
    grad_y_flip = np.zeros((grad_y.shape))
    
    #variable for slicing the img matrix according to the kernal size, and increment it by 1 after each iteration
    row_slice= filter_size[0]
    
    #flip the filters in x and y direction but as they are 
    #gaussian so just flip them in the opposite axis only, i.e fx in y direction fy in x direction
    pos = 0
    for x in grad_x:
        r = x[::-1]
        grad_x_flip[pos] =r
        pos+=1
        
    #print(grad_x_flip)
    grad_y = np.flipud(grad_y)
    
    #1:Slice the image according to filter size
    for i in range(row):
        col_slice= filter_size[1]
        
        for j in range(col):
            img_slice = img[i:row_slice , j:col_slice]
            
            #2: convole and add the value into the corresonding derivative images
            fx[i+1, j+1] = sum(sum(img_slice*grad_x_flip))
            fy[i+1, j+1] = sum(sum(img_slice*grad_y))
            
            #increment column slice size by 1
            col_slice = col_slice+1
            
        #increment the row slice by 1
        row_slice = row_slice+1

    return fx,fy
```

# Compute gradiant magnitude.......M = sqrt(fx^2 + fy^2)

```python
def gradiant_magnitude(fx,fy):
    scale = 255
    M = np.sqrt(fx**2 + fy**2)/scale
    return M
```

# Compute Gradiant Direction  theta = arctan(fy/fx)

```python
def gradiant_direction(fx,fy):
    dimention = fx.shape
    grad_direction = []
    fy = fy.flatten().tolist()
    fx = fx.flatten().tolist()
    
    for x,y in zip(fx,fy):
        theta = m.atan2(y,x)
        
        #change to degrees and add 180
        degrees = (theta*180/m.pi) +180
        grad_direction.append(degrees)
        
    grad_direction = np.reshape(grad_direction, dimention)
    
    return grad_direction
```

# Non-Maxima Suppression

```python
def convert_to_RGB(quantized_img):
    row,col = quantized_img.shape
    rgb = np.zeros((row,col,3))
    
    red = [144, 12, 63]
    green = [218, 247, 166]
    blue = [0, 255, 255]
    jade = [0, 163, 108]
    yellow = [255, 195, 0]
    
    for i in range(row):
        for j in range(col):
            if quantized_img[i,j] == 0:
                rgb[i,j] = blue
            
            elif quantized_img[i,j] == 1:
                rgb[i,j] = green
             
            elif quantized_img[i,j] == 2:
                rgb[i,j] = red
                
            elif quantized_img[i,j] == 3:
                rgb[i,j] = yellow
                
            else:rgb[i,j] = jade
                
    return rgb
              
```

```python
def NMS(grad_dir,grad_mag):
    quantize = []
    row,col = grad_dir.shape
    grad_dir_list = grad_dir.flatten().tolist()
    grad_mag_nms = np.zeros((grad_mag.shape))
    
    #1: quantization of gradiant directions
    for val in grad_dir_list:
        if val == 0:
            quantize.append(0)
            
        elif 0<val<22.5 or 157.5<=val<202.5 or 337.5<=val<=360:
            quantize.append(1)
            
        elif 22.5<=val<67.5 or 202.5<=val<247.5:
            quantize.append(2)
        
        elif 67.5<=val<112.5 or 247.5<=val<292.5:
            quantize.append(3)
            
        elif 112.5<=val<157.5 or 292.5<=val<337.5:
            quantize.append(4)
            
    quantize = np.reshape(quantize,(row,col))
    
    #2: Non maximum supression: look for the edge direction
    #   from the quatized matrix and change the value in M acordingly
    for i in range(1,row-1):
        for j in range(1,col-1):
            if grad_mag[i,j] != 0:
                #check the direction of gradiant from quantized matrix
                if quantize[i,j]==quantize[i+1,j] and quantize[i,j]==quantize[i-1,j]:
                    if grad_mag[i,j]>grad_mag[i+1,j] and grad_mag[i,j]>grad_mag[i-1,j]:
                        #print('value converted to zero in north south ')
                        grad_mag_nms[i,j] = grad_mag[i,j]
                        
                elif quantize[i,j]==quantize[i,j+1] and quantize[i,j]==quantize[i,j-1]:
                    if grad_mag[i,j]>grad_mag[i,j+1] and grad_mag[i,j]>grad_mag[i,j-1]:
                        #print('value converted to zero in east west')
                        grad_mag_nms[i,j] = grad_mag[i,j]
                                                     
                elif quantize[i,j]==quantize[i-1,j+1] and quantize[i,j]==quantize[i+1,j-1]:
                    if grad_mag[i,j]>grad_mag[i-1,j+1] and grad_mag[i,j]>grad_mag[i+1,j-1]:
                        #print('value converted to zero in +ve diagnol')
                        grad_mag_nms[i,j] = grad_mag[i,j]
                        
                elif quantize[i,j]==quantize[i-1,j-1] and quantize[i,j]==quantize[i+1,j+1]:
                    if grad_mag[i,j]>grad_mag[i-1,j-1] and grad_mag[i,j]>grad_mag[i+1,j+1]:
                        #print('value converted to zero in -ve diagnol')
                        grad_mag_nms[i,j] = grad_mag[i,j]
   
   
    
    return convert_to_RGB(quantize), grad_mag_nms
        
```

# Hysteresis thresholding

```python
def recurs_n(img,row,col,visited,hest_img,T1):
    for i in range(row-1,row+1):
        for j in  range(col-1,col+1):
            if visited[i,j]==0:
                visited[i,j] = 1
                if img[i,j]>T1:
                    hest_img[i,j] = 1
                    recurs_n(img,i,j,visited,hest_img,T1)
    return 
```

```python
def hys_threshold(img,Th,T1):
    #padd the image with zeros
    img = np.pad(img, ((1,1),(1,1)), 'constant')
    
    visited = np.zeros((img.shape))
    hest_img = np.zeros((img.shape))
    
    row,col = img.shape
    
    for i in range(1,row-1):
        for j in range(1,col-1):
            if visited[i,j] == 0:
                visited[i,j] = 1
                if img[i,j] > Th:
                    hest_img[i,j] = 1
                    recurs_n(img,i,j,visited,hest_img,T1)
    return hest_img
```

```python
#genrate gradiants in both x and y direction with all three sigma value filters
def gen_fx_fy(img, Gx_5, Gy_5, Gx_1, Gy_1, Gx_2, Gy_2):
    #1: convert to gray scale
    gray = convert_to_greyScale(img)

    #2:blur the image to get wider edges
    blur = smooth_img(gray)

    #3: convolve with gaussian mask of 
    #sigma= 0.5
    fx_5, fy_5 = convolution(blur,Gx_5,Gy_5)

    #sigma= 1
    fx_1 ,fy_1 = convolution(blur,Gx_1,Gy_1)

    #sigma= 2.0
    fx_2, fy_2 = convolution(blur,Gx_2,Gy_2)
    
    return [fx_5, fy_5, fx_1, fy_1, fx_2, fy_2]
```

```python
#4:compute gradiant magnitude
#list passed is in the form [fx_5, fy_5, fx_1, fy_1, fx_2, fy_2]
def compute_grad_mag(lis):
    # fx_5 and fy_5
    mag_5 = gradiant_magnitude(lis[0],lis[1])
    # fx_1 and fy_1
    mag_1 = gradiant_magnitude(lis[2],lis[3])
    # fx_2 and fy_2
    mag_2 = gradiant_magnitude(lis[4],lis[5])
    
    return [mag_5,mag_1,mag_2]
```

```python
#5: compute gradiant direction at all 3 sigma values variables
#list passed is in the form [fx_5, fy_5, fx_1, fy_1, fx_2, fy_2]
def compute_gradDir(lis):
    grad_dir_5 = gradiant_direction(lis[0], lis[1])
    grad_dir_1 = gradiant_direction(lis[2], lis[3])
    grad_dir_2 = gradiant_direction(lis[4], lis[5])
    return [grad_dir_5, grad_dir_1, grad_dir_2]
```

```python
#6: apply non maximum supression for all 3 sigma values variables
#list passed is in the form [mag_5, mag_1, mag_2] and [grad_dir_5, grad_dir_1, grad_dir_2]
def compute_NMS(mag_list, dir_list):
    quantized_5 , grad_mag_nms_5 = NMS(dir_list[0] , mag_list[0])
    quantized_1 , grad_mag_nms_1 = NMS(dir_list[1] , mag_list[1])
    quantized_2 , grad_mag_nms_2 = NMS(dir_list[2] , mag_list[2])
    return [quantized_5 ,quantized_1, quantized_2 ], [grad_mag_nms_5, grad_mag_nms_1, grad_mag_nms_2]
```

# Mathematical Part: surface plot of 1st and 2nd derivative of gaussian

```python
#calculate 1st and 2nd derivates of gaussian
def first_derivative(x,y,sigma):
    fx = np.exp(-(x**2 + y**2)/(2*sigma**2)) * (x/(sigma**2))
    fy = np.exp(-(x**2 + y**2)/(2*sigma**2)) * (y/(sigma**2))
    return fx,fy

def sec_derivative(x,y,sigma):
    fxx = np.exp(-(x**2 + y**2)/(2*sigma**2)) * ((x**2-sigma**2)/sigma**4)
    fyy = np.exp(-(x**2 + y**2)/(2*sigma**2)) * ((y**2-sigma**2)/sigma**4)
    return fxx,fyy
```

```python
def show_and_save_surfaceplot(x,y,z,title):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x,y,z, cmap=cm.jet)
    plt.title(title)
    fig.savefig(title+'.jpg')
    plt.show()
```
